```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = '../data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
app_train.plot.scatter(x='AMT_CREDIT', y='AMT_INCOME_TOTAL', c='AMT_ANNUITY')

import seaborn as sns
sns.pairplot(x_vars=['AMT_CREDIT'], y_vars=['AMT_INCOME_TOTAL'], data=app_train, hue="CODE_GENDER")

print('AMT_CREDIT_AVERAGE')
print(app_train['AMT_CREDIT'].mean())
print('AMT_CREDIT_STDEV')
print(app_train['AMT_CREDIT'].std())

#groups=app_train.groupby('CODE_GENDER')
#for name, group in groups:
#    group.plot.scatter(x='AMT_CREDIT', y='AMT_INCOME_TOTAL', label=name)
```
